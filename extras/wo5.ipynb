```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading the excel file
data = pd.read_excel("preprocessed_data.xlsx")


```

```python
#Question 5
from scipy import stats
slope, intercept, regress, p, std_err = stats.linregress(data['hand_length_left'],data['hand_length_right'])
plt.scatter(data["hand_length_left"],data["hand_length_right"])

plt.plot(data["hand_length_left"],intercept + slope*data['hand_length_left'],label = 'best_fit_line',color='black')
print(slope)
plt.show()
```

```python
#creating the box plot
count_row = data.shape[0] # counts the number of rows
data["diff"] = 0
for i in range(count_row):
    data["diff"][i] = abs(data["hand_length_left"][i] - data["hand_length_right"][i])

#Plotting the box plot
plt.boxplot(data["diff"])
plt.show()
```

```python
#below Q1 − 1.5×IQR or above Q3 + 1.5×IQR

q1 = np.percentile(data["diff"],25)
q3 = np.percentile(data["diff"], 75)  
inter_quartile = q3 - q1

#now dropping the outliers

temp1 = q1 - 1.5*inter_quartile
temp2 = q3 + 1.5*inter_quartile

data.drop(data.loc[(data['diff']<temp1) |(data["diff"]>temp2)].index,inplace =True)
```

```python
slope, intercept, regress, p, std_err = stats.linregress(data['hand_length_left'],data['hand_length_right'])
plt.scatter(data["hand_length_left"],data["hand_length_right"])

plt.plot(data["hand_length_left"],intercept + slope*data['hand_length_left'],label = 'best_fit_line',color='black')
print(slope)
plt.show()
```
